```python
#Exercise 6-1
import matplotlib.pyplot as plt
imglist=[]
size = 100
for y in range(size):
  xlist=[]
  for x in range(size):
    val=(x*y%50) #this determines how the brightness of the color is determined
    xlist.append(val)
  imglist.append(xlist)
plt.imshow(imglist, cmap='coolwarm')
```

```python
#Exercise 6-2
import matplotlib.pyplot as plt
imglist = []
size = 100
for y in range(size):
  xlist=[]
  for x in range(size):
    if y>=20 and x>=20 and y<=80 and x<=80: #creating the square in white
      val=1.0 #this determines how the brightness of the color is determined
    else:
      val=0 #other than the square comes black
    xlist.append(val)
  imglist.append(xlist)
plt.imshow(imglist,cmap = 'gray')
plt.show()
```

```python
#Exercise 6-3
templist=[]
rf = open('TokyoTemp1941-2020.csv','r') #opening the file
for line in rf:
  striped = line.strip() #this is removing the line feed code
  averages=striped.split(',') #splitting the data between ','
  eylist=[]
  for average in averages: #adding each element from the line into the list
    average=float(average)
    eylist.append(average)
  templist.append(eylist)
print(templist)

import matplotlib.pyplot as plt
plt.imshow(templist, cmap='coolwarm')
plt.show()
rf.close()
```

```python
#Exercise 6-5(a)
templist=[]
rf = open('TokyoTemp1941-2020.csv','r') #opening the file
for line in rf:
  striped = line.strip() #this is removing the line feed code
  averages=striped.split(',') #splitting the data between ','
  eylist=[]
  for average in averages: #adding each element from the line into the list
    average=float(average)
    eylist.append(average)
  templist.append(eylist)

eylist=[]
for i in range(80):
  sum=0
  for element in templist[i]:
    sum+=element
  eylist.append(float(sum/12)) #the list of the average temeperature of each year

import matplotlib.pyplot as plt #plotting a line graph of the eylist
plt.plot(eylist)
plt.xlabel('year')
plt.ylabel('temperature')
plt.show()
rf.close()
```

```python
#Exercise 6-5(b)
templist=[]
rf = open('TokyoTemp1941-2020.csv','r') #opening the file
for line in rf:
  striped = line.strip() #this is removing the line feed code
  averages=striped.split(',') #splitting the data between ','
  eylist=[]
  for average in averages: #adding each element from the line into the list
    average=float(average)
    eylist.append(average)
  templist.append(eylist)

emlist=[]
for m in range(12): #m for the month
  sum=0
  for y in range(80): #y for the year
    sum+=templist[y][m] #adding corresponding month of every year to the sum
  emlist.append(float(sum/80)) #calculating the average of each month over the years

import matplotlib.pyplot as plt #plotting a line graph of the emlist
plt.plot(emlist)
plt.xlabel('month')
plt.ylabel('temperature')
plt.show()
rf.close()
```

```python
#Exercise 6-7
imglist=[]
rf=open('240x160data.csv','r')
for line in rf:
  striped = line.strip() #this is removing the line feed code
  numbers=striped.split(',') #splitting the data between ','
  xylist=[]
  for number in numbers:
    i=float(number)
    xylist.append(1-i) #all the elements are converted to 1-x
  imglist.append(xylist) #creating a list of elemennts in the data file
import matplotlib.pyplot as plt
plt.imshow(imglist, cmap='gray')
rf.close()
```

```python
#Exercise 6-8
imglist=[]
imglist2=[]
rf=open('240x160data.csv','r')
for line in rf:
  striped = line.strip() #this is removing the line feed code
  numbers=striped.split(',') #splitting the data between ','
  xylist=[]
  for number in numbers:
    i=float(number)
    xylist.append(i)
  imglist.append(xylist) #creating a list of elemennts in the data file

imglist8=[]
#creat the 8x8 square
for x in range(0,240,8): #looping through row from 0 to 240 every 8th element
  square=[]
  for y in range(0,120,8): #loop through columnsfrom 0 to 240 every 8th element
    sum=0
    count=0
    for i in range(x, x+8): #loop through rows within the 8x8
      for l in range(y, y+8): #loop through collumns withint the 8x8
        sum+=imglist[i][l] #sum of all the elements within the 8x8 square
        count+=1
    av=sum/count #average of the 8x8 sqaure
    square.append(av)
  imglist8.append(square)

import matplotlib.pyplot as plt
plt.imshow(imglist8, cmap='gray')
rf.close()
```
