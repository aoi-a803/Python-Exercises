EX8-1: Referring to p.10 Prg. 2 (which calculates the sum of numbers up to n), create a recursive function factorial() that calculate the factorial of the argument n (n*(n1)*...*2*1=n!).
• Note: Since 0! is equal to 1, when n == 0, the function should return 1.
• Example: For n=3, it will compute 3 * 2 * 1 = 6.

```python
#Exercise 8-1
def cal(n):
  if n == 0:
    return 1 #outputting 1 when n==0
  else:
    kaijo = n * cal(n-1) #recalling the calculation until the previous number and multiplying by it
    return kaijo

print(cal(4))
```

EX8-3: Create a function combi(n,r) that calculates the number of ways to select r distinct items from n distinct items.
• Arguments: n, r
• Returns: The number of ways to select r distinct items from n distinct items (not the actual combination)
• Example: combi(5,3) calculates the number of ways to select 3 distinct items from 5 items like a, b, c, d, e.
The possible combinations are (a, b, c), (a, b, d), (a, b, e), (a, c, d),..., which results in 10 patterns.
Create the recursive combi(n,r) function based on the following rules. Print the result for n=5, r=3 :
• Rule 1: If r = 0 or r = n, then combi(n, r) = 1
• Rule 2: Otherwise, use the relation: combi(n, r) = combi (n-1, r) + combi (n-1, r-1)
• Note: Do not use the additional rules given below for the implementation.

```python
#Exercise 8-3
def combi(n,r):
  if r==0 or r==n: #while the condition is True, the output would be '1'
    return 1
  else:
    sumpattern= combi (n-1, r) + combi (n-1, r-1) #summing up the number of combinations of r from n; recalling the looping calculation
    return sumpattern

print(combi(5,3))
```

EX8-11: Based on the program from EX8-10, create the following function recls_sizedict().
• Arguments: Path to the directory
• Return value: A dictionary with information about all files within the specified directory and its subdirectories.
  • Key: File path
  • Value: File size
• Execute recls_sizedict() with the path “../datalab/web” and print the results.
• Instead of printing the dictionary as a whole, use .items() to display each key-value pair on a separate line.

```python
#Exercise 8-11
import os

def recls_sizedict(directory):
    newdict = {}
    for entry in os.scandir(directory):
        if entry.is_file():
            newdict[entry.path] = entry.stat().st_size  #setting file name as the key and its size as the value
        elif entry.is_dir():
            newdict.update(recls_sizedict(entry.path))  #adding the keys and values extracted from the subdirectory
    return newdict

for file_name, file_size in recls_sizedict("../datalab/web").items(): #printing each file name and its size on every other line
    print(f"{file_name}: {file_size}")
```
