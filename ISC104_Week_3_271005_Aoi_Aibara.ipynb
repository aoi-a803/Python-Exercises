EX3-10: Coupon collector problem, Gacha comp problem
Simulate rolling a 6-sided die and keep rolling until all numbers from 1 to 6 have appeared at least once.
Track how many rolls it takes. Repeat this 100 times and calculate the average number of rolls it takes.

```python
#Exercise 3-10
import random
count=0
dice_numbers=set()

while len(dice_numbers)<6: #since set() doesn't allow duplication, it will be less than 6 when not all numbers are rolled
  dice=random.randint(1,6)
  dice_numbers.add(dice) #adding to the list of rolled dice
  count+=1 #keeping track of how many times it rolled
print(count)
```

```python
#Exercise 3-10 repeating 100 times
import random
count=0
dice_numbers=set()

while len(dice_numbers)<6: #since set() doesn't allow duplication, it will be less than 6 when not all numbers are rolled
  dice=random.randint(1,6)
  dice_numbers.add(dice) #adding to the list of rolled dice
  count+=1 #keeping track of how many times it rolled

total=0
for _ in range(100):
  total+=count #Calculating the sum of numbers it took the dice to get all the numbers out
average = total/100
print("The average number to see all the numbers is ", average)
```
