EX 10-4:
Creat the function

Arguments: two lists (same number of elements, elements are
numeric)

Returns: the sum of two lists multiplied by the value of the same
index productsum([6,4,2],[1,3,5]) -> 28 (=6*1 + 4*3 + 2*5 )

```python
#Exercise 10-4
def productsum(list1, list2):
  sum=0 #define the sum
  for n in range(len(list1)): #for every element in the list
    sum +=list1[n]*list2[n] #sum up the multiplication of each nth element
  return sum
print(productsum([6,4,2],[1,3,5]))
```

EX10-5: Give the combination of the two menu items with the highest total
calorie value. Provide both the combination and the total calories. (Each menu
can only be chosen once)


```python
#Exercise 10-5
import itertools
kcaldict = {'udon':300, 'salad':150, 'gyudon':500, 'pasta':450}

def high_cal():
  kcallist=list(kcaldict.keys()) #converting the dictionary into a list
  combi=list(itertools.combinations(kcallist, 2)) #making combinations of two from the list
  highest = 0 #setting the initial number
  highest_combi=None #asigning it to begin with nothing

  for c in combi: #to investigate all the cominations
    total = kcaldict[c[0]]+kcaldict[c[1]] #the total of within the combination
    if total > highest: #evaluating the highest total sum of the combination of each combination
      highest = total #replacing the highest each turn when the if() is True
      highest_combi= c #when if() is True, we want to know which combination that is

  return highest, highest_combi #outputting the two variables

highestcal, highestcombi = high_cal() #asigning the names to the valuables outputted from high_cal()
print(f"the highest total sum of the combinations is: {highestcal}") #calling out for highestcal
print (f"the combination is: {highestcombi}") #calling out for highest combi
```

calorie content and cost less than 800 yen. (Each menu can only be chosen
once)

EX 10-8: Write a program to find the longest sequence of the same consecutive elements in a list. Print the number of elements and the element.

```python
#Exercise 10-8
def sequency(lst):
  high_count=0
  when = None
  current_count =1
  current = lst[0]

  for i in range(1,len(lst)):
    if lst[i]==lst[i-1]:
      current_count+=1
    else:
      if current_count>high_count:
        high_count=current_count
        when = current
      current_count=1 #reset the counting
      current=lst[i] #update the element

  if current_count>high_count:
    high_count=current_count
    when = current

  return high_count, when

lst = [1, 1, 2, 2, 2, 3, 3, 1, 1, 1, 1]
high_count, when = sequency(lst)
print(f"the longest sequence of the same number in the list is {when} which appeared {high_count} times")
```

EX 10-9: Calculator for Expressions with Two Numbers (Cont’d)

```python
#Exercise 10-9
class Calculator:
  #below is creating methods in the application to navigate each implementation depending on what is called
  def add(self, a, b):
    return a+b

  def sub(self, a, b):
    return a-b

  def multiply(self, a, b):
    return a*b

  def division(self, a, b):
    if b==0:
      return 'Cannot divide by zero'
    return a/b

  def is_number(self, s):
    try: #testing the following statement
      float(s) #if the string can be converted to float,
      return True
    except:
      return False

  def evaluate(self, expression):
    operators = ['+', '-', '*', '/']
    for i in operators:
      splitted = expression.split(i)
      if len(splitted)==2:
        splitted1, splitted2 = splitted

        if self.is_number(splitted1) and self.is_number(splitted2): #making sure that the values are numbers by running is_number
          splitted1 = float(splitted1) #if they are, convert it to float
          splitted2 = float(splitted2)

          #below assigns the methods for each depending on what operator is included
          if i =='+':
            return self.add(splitted1, splitted2)
          elif i == '-':
            return self.subtract(splitted1, splitted2)
          elif i == '*':
            return self.multiply(splitted1, splitted2)
          elif i == '/':
            return self.division(splitted1, splitted2)
    return 'no operator or multiple operators)'

calc=Calculator()

print(calc.evaluate("1+2"))
print(calc.evaluate("5 * 4"))
print(calc.evaluate("10/2"))
print(calc.evaluate("5/0"))
print(calc.evaluate("1+"))
```
